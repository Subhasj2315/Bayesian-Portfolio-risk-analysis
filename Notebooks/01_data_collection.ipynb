```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import yfinance as yf
ticker = yf.Ticker("RELIANCE.NS")
data_reliance  = ticker.history(period ="1y" )
print(data_reliance.head())
```

```python
stocks = {
    "TCS":"TCS.NS",
    "Infosys":"INFY.NS",
    "Reliance":"RELIANCE.NS",
    "HDFC":"HDFCBANK.NS",
    "ITC":"ITC.NS"
} # here we are taking Data from 5 companies for 5 years
```

```python
data = {} #empty dictionary.

for name,ticker in stocks.items():

    df = yf.download(
        ticker,
        start="2019-01-01",
        end="2024-12-31"
    )

    data[name] = df
```

```python
data["ITC"]
```

```python
for name,df in data.items():
       print(data[name].isnull().sum()) # so no missing values
```

```python
for name,df in data.items():
       print(df.columns)
```
